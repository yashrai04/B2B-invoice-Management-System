# Milestone-3 (Feature Engineering and Selection)

--> 1. Check for the outliers in the “amount_in_usd” column and replace the outliers with appropriate values, discussed in the sessions.

```python
df['amount_in_usd'].skew()
```

```python
# Calculate the quartiles and IQR
df["amount_in_usd"] = pd.to_numeric(df["amount_in_usd"], errors='coerce')

pect_25 = df["amount_in_usd"].quantile(0.25)
pect_75 = df["amount_in_usd"].quantile(0.75)
iqr = pect_75 - pect_25

# Define the upper and lower limits for outliers
upper_bound = pect_75 + 1.5 * iqr
lower_bound = pect_25 - 1.5 * iqr

print(pect_25)
print(pect_75)
```

```python
df["amount_in_usd"]=np.where(df['amount_in_usd']>upper_bound,df["amount_in_usd"].median(),np.where(df['amount_in_usd']<lower_bound,df["amount_in_usd"].median(),df['amount_in_usd']))
df
```

--> 2 Label encoding or One hot Encoding on all the categorical columns 

```python
# Label encoding or One hot Encoding on all the categorical columns 
from sklearn.preprocessing import LabelEncoder

categorical_data = ['DISTRIBUTION_CHANNEL', 'DIVISION', 'PURCHASE_ORDER_TYPE', 'CREDIT_CONTROL_AREA', 'ORDER_CURRENCY']

label_encoder = LabelEncoder()

for data in categorical_data:
    df[data] = label_encoder.fit_transform(df[data])

print(df)
```

3 Log Transformations on continuous columns 

```python
# Log Transformations on continuous columns 
# Replacing ',' in RELEASED_CREDIT_VALUE column to '.'
df['RELEASED_CREDIT_VALUE'] = df['RELEASED_CREDIT_VALUE'].astype(str)  # Convert the column to string type
df['RELEASED_CREDIT_VALUE'] = df['RELEASED_CREDIT_VALUE'].apply(lambda x: x.replace(',', '.'))  # Replace ',' with '.'
df['RELEASED_CREDIT_VALUE'] = df['RELEASED_CREDIT_VALUE'].apply(lambda x: x.replace('-', ''))  # Replace '-' with '.'
df['RELEASED_CREDIT_VALUE'] = df['RELEASED_CREDIT_VALUE'].astype(float)  # Convert the column back to float type

# Three columns are continuous RELEASED_CREDIT_VALUE, ORDER_AMOUNT , amount_in_usd 
df['RELEASED_CREDIT_VALUE'] = np.log1p(df['RELEASED_CREDIT_VALUE'])
df['ORDER_AMOUNT'] = np.log1p(df['ORDER_AMOUNT'])
df['amount_in_usd'] = np.log1p(df['amount_in_usd'])
df
```

-->4 Try to extract new features by grouping existing columns 

```python
# Total Released Credit Value by Sales Organization:
# Grouping by the SALES_ORG and RELEASED_CREDIT_VALUE column, you can calculate the total released credit value for each sales organization and avg. This can provide info about the credit capacity of different sales teams
df['total_credit_value_by_sales_org'] = df.groupby('SALES_ORG')['RELEASED_CREDIT_VALUE'].transform('sum')
df['total_credit_value_by_sales_org']
```

```python
# Maximum Order Amount by Company Code:
# Grouping by the COMPANY_CODE and amount_in_usd column, you can find the maximum order amount within each company code. This can help identify the company with the highest order amount.
df['max_order_amount_by_company_code'] = df.groupby('COMPANY_CODE')['amount_in_usd'].transform('max')
df['max_order_amount_by_company_code']
```

```python
# Average order amount by division:
# the values of the 'average_order_amount_by_division' feature, you can identify divisions with higher or lower order amounts compared to the overall average.
df['average_order_amount_by_division'] = df.groupby('DIVISION')['amount_in_usd'].transform('mean')
df['average_order_amount_by_division']
```

```python
# Requested Delivery Date Difference
# by subtracting
# ORDER_CREATION_DATE from REQUESTED_DELIVERY_DATE we can identify the requested_delivery_date_diff which helps to identify days taken to deliver an order 
df['requested_delivery_date_diff'] = pd.to_datetime(df['REQUESTED_DELIVERY_DATE']) - pd.to_datetime(df['ORDER_CREATION_DATE'])
df['requested_delivery_date_diff'] = df['requested_delivery_date_diff'].dt.days
df['requested_delivery_date_diff']
```

```python
def difference_in_days(melt, lags, ffday, customer_id_col, create_date_col, net_amount_col):
    for i in range(ffday, lags+1):
        melt['Last-'+str(i)+'day_Sales'] = melt.groupby([customer_id_col])[net_amount_col].shift(i)
    
    melt = melt.reset_index(drop = True)
    
    for i in range(ffday, lags+1):
        melt['Last-'+str(i)+'day_Diff']  = melt.groupby([customer_id_col])['Last-'+str(i)+'day_Sales'].diff()
    melt = melt.fillna(0)
    return melt
```

# Milestone-4 (ML Models and Evaluations)

spliting data into training and testing and selecting input features

```python
from sklearn.model_selection import train_test_split
input_features =[
    'CUSTOMER_ORDER_ID',
    'SALES_ORG',
    'COMPANY_CODE',
    'ORDER_CREATION_TIME',
    'SOLD_TO_PARTY',
    'CUSTOMER_NUMBER',
    'DIVISION',
    'DISTRIBUTION_CHANNEL',
    'PURCHASE_ORDER_TYPE',
    'COMPANY_CODE',
    'CREDIT_CONTROL_AREA',
    'ORDER_CURRENCY',
    'RELEASED_CREDIT_VALUE',
    'CREDIT_STATUS',
    'total_credit_value_by_sales_org',
    'max_order_amount_by_company_code',
    'average_order_amount_by_division',
    'requested_delivery_date_diff',
]

output_feature = 'amount_in_usd'

X = df[input_features]
y = df[output_feature]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)

print("Training set shape - Input features:", X_train.shape)
print("Training set shape - Output feature:", y_train.shape)
print("Testing set shape:", X_test.shape, y_test.shape)
```

#  LinearRegression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.impute import SimpleImputer
import numpy as np

imputer = SimpleImputer(strategy='mean')

X_train_imputed = imputer.fit_transform(X_train)
X_test_imputed = imputer.transform(X_test)

model = LinearRegression()
model.fit(X_train_imputed, y_train)

train_predictions = model.predict(X_train_imputed)
test_predictions = model.predict(X_test_imputed)

train_mse = mean_squared_error(y_train, train_predictions)
test_mse = mean_squared_error(y_test, test_predictions)

train_rmse = np.sqrt(train_mse)
test_rmse = np.sqrt(test_mse)

train_r2 = r2_score(y_train, train_predictions)
test_r2 = r2_score(y_test, test_predictions)

print("Train MSE:", train_mse)
print("Test MSE:", test_mse)
print("Train RMSE:", train_rmse)
print("Test RMSE:", test_rmse)
print("Train R-squared:", train_r2)
print("Test R-squared:", test_r2)
```

Ploting Linear Regression Graph

```python
import matplotlib.pyplot as plt
plt.scatter(y_train, train_predictions, color='blue', label='Training Data')
plt.scatter(y_test, test_predictions, color='red', label='Testing Data')
plt.plot([y.min(), y.max()], [y.min(), y.max()], color='black', linestyle='--', label='Ideal')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Linear Regression Predictions')
plt.legend()
plt.show()
```

# Hyperparameter tuning for Xgboast

```python
import xgboost as xgb
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

xgb_model = xgb.XGBRegressor()

param_grid = {
    'max_depth': [3, 5, 7],
    'learning_rate': [0.1, 0.01, 0.001],
    'n_estimators': [100, 500, 1000],
}

random_search = RandomizedSearchCV(xgb_model, param_grid, cv=5, scoring='neg_mean_absolute_error')
random_search.fit(X_train_imputed, y_train)

best_model = random_search.best_estimator_
best_params = random_search.best_params_
best_model.fit(X_train_imputed, y_train)
y_train_pred = best_model.predict(X_train_imputed)

mae_train_xg = mean_absolute_error(y_train, y_train_pred)
mse_train_xg = mean_squared_error(y_train, y_train_pred)
rmse_train_xg = mean_squared_error(y_train, y_train_pred, squared=False)
r2_train_xg = r2_score(y_train, y_train_pred)

y_test_pred = best_model.predict(X_test_imputed)

mae_test_xg = mean_absolute_error(y_test, y_test_pred)
mse_test_xg = mean_squared_error(y_test, y_test_pred)
rmse_test_xg = mean_squared_error(y_test, y_test_pred, squared=False)
r2_test_xg = r2_score(y_test, y_test_pred)

print("Best Hyperparameters:", best_params)
print()
print("Training Data:")
print("Mean Absolute Error (MAE):", mae_train_xg)
print("Mean Squared Error (MSE):", mse_train_xg)
print("Root Mean Squared Error (RMSE):", rmse_train_xg)
print("R-squared (R2):", r2_train_xg)
print()

print("Test Data:")
print("Mean Absolute Error (MAE):", mae_test_xg)
print("Mean Squared Error (MSE):", mse_test_xg)
print("Root Mean Squared Error (RMSE):", rmse_test_xg)
print("R-squared (R2):",r2_test_xg)
```

After Hyperparameter tuning R-squared Score of the model is reduced to 0.6823157299339059 from 0.7173569210518731 which inticates that our model cannot be futher tunned to get more R2 score and hence the Xgboast model with R2 score of 0.7173569210518731 is the best selected model
